```python
import os
import pandas as pd
```

```python
fifteen_path = os.path.join('data','csv','happiness_2015.csv')
sixteen_path = os.path.join('data','csv','happiness_2016.csv')
seventeen_path = os.path.join('data','csv','happiness_2017.csv')
# country_path = os.path.join('Data','csv','countries.csv')
```

```python
happiness_15_df = pd.read_csv(fifteen_path)
happiness_16_df = pd.read_csv(sixteen_path)
happiness_17_df = pd.read_csv(seventeen_path)
# country_df = pd.read_csv(country_path)
```

```python
list_2015 = []
for country in happiness_15_df['Country']:
    list_2015.append(country)
list_2016 = []
for country in happiness_16_df['Country']:
    list_2016.append(country)
list_2017 = []
for country in happiness_17_df['Country']:
    list_2017.append(country)
```

```python
print(list_2015)
print(list_2016)
print(list_2017)
```

```python
country_list = []
def list_country(l):
    for country in l:
        if country in country_list:
            continue
        else:
            country_list.append(country)
```

```python
list_country(list_2015)
list_country(list_2016)
list_country(list_2017)
```

```python
country_list.sort()
```

```python
country_list
```

```python
country_id = []
counter = 1
for country in country_list:
    country_id.append(counter)
    counter += 1
```

```python
country_data = {'country':country_list, 'country_id':country_id}
```

```python
country_data
```

```python
country_df = pd.DataFrame.from_dict(country_data)
```

```python
country_df.to_csv(os.path.join('Data','csv','countries.csv'))
```

```python
country_id_dict = dict()
for index,row in country_df.iterrows() :
    country_name = row['country']
    country_id = row['country_id']
    country_id_dict[country_name] = country_id
```

```python
country_id_dict
```

```python
happiness_15_df['country_id'] =""
happiness_16_df['country_id'] =""
happiness_17_df['country_id'] =""
```

```python
def add_country_id(df):
    for index,row in df.iterrows() :
        try:
            df.loc[index,'country_id'] = country_id_dict.get(row['Country'])
        except:
            continue
```

```python
add_country_id(happiness_15_df)
add_country_id(happiness_16_df)
add_country_id(happiness_17_df)
```

```python
happiness_15_df.sort_values(by=['Country'])
happiness_16_df.sort_values(by=['Country'])
happiness_17_df.sort_values(by=['Country'])
```
